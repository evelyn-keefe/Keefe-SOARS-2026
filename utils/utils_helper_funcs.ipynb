```python
from pathlib import Path
import xarray as xr
import netCDF4
```

```python
# Helper functions
def inspect(ds):
    # Check that the dataset has variables necessary to compute pressure coords
    try:
        ds['P0']
        ds['hyai']
        ds['hybi']
        ds['PS']
    # If not, raise an error
    except:
        raise KeyError('This dataset does not have all components needed to handle the pressure coordinates')
        
    return None

def create_big_ds(path):
    files = []
    p = Path(path)
    for file in p.iterdir():
        if str(file.name) == '.ipynb_checkpoints':
            file = next(p.iterdir())
        files.append(file)
    big_ds = xr.open_mfdataset(files, compat='override', coords='minimal', engine='netcdf4')
    return big_ds
```
