```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
# Experimenting with importing shapefiles and emissions
shapes = '/glade/campaign/acom/MUSICA/emis/htap3/regional_mask_0.1deg.nc'
ds_mask = xr.open_dataset(shapes)
ds_mask
```

```python
# Extract the Australian and North American regions
lat = ds_mask.latitude
lon = ds_mask.longitude
reg = ds_mask.region_mask
BNA = reg.sel(region = 'BNA') # Boreal North America
TNA = reg.sel(region = 'TNA') # Temperate North America
ANZ = reg.sel(region = 'ANZ') # Australia and New Zealand
BA = reg.sel(region = 'BA') # Boreal Asia
Eur = reg.sel(region = 'Eur') # Europe
SA = reg.sel(region = 'SA') # South Asia
AME = reg.sel(region = 'AME') # Africa and Middle East
SAm = reg.sel(region = 'SAm') # South America
reg_list = [BA, BNA, Eur, TNA, SA, AME, SAm, ANZ]
```

```python
BA_only = BA.where(BA.values == 1)
BNA_only = BNA.where(BNA.values == 1)
Eur_only = Eur.where(Eur.values == 1)
TNA_only = TNA.where(TNA.values == 1)
SA_only = SA.where(SA.values == 1)
SAm_only = SAm.where(SAm.values == 1)
ANZ_only = ANZ.where(ANZ.values == 1)
only_list = [BA_only, BNA_only, Eur_only, TNA_only, SA_only, SAm_only, ANZ_only]
```
